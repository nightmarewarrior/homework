```python
from pymystem3 import Mystem
from collections import Counter
```

```python
text = """
Античит-движки для игр — это ещё одна крупная область Windows, которая использует драйверы на уровне ядра. Microsoft ведёт переговоры с разработчиками игр о том, как сократить объём использования ядра. Riot Games сообщила Microsoft о готовности следовать потенциальным изменениям безопасности Windows и сократить влияние на ядро ОС.

Летом этого года Microsoft готовится к выпуску обновления Windows, которое будет включать новую функцию Quick Machine Recovery. Функция предназначена для быстрого восстановления системы после сбоя. Ранее Microsoft сообщила о планах сменить синий экран смерти на чёрный.
"""
```

```python
m = Mystem()
parsed = m.analyze(text)
```

```python
filtered = [token for token in parsed if token.get('analysis')]
```

```python
pos_tags = [
    token['analysis'][0]['gr'].split(',')[0].split('=')[0]
    for token in filtered
]
```

```python
trigrams = [
    (pos_tags[i], pos_tags[i+1], pos_tags[i+2])
    for i in range(len(pos_tags) - 2)
]
trigram_counts = Counter(trigrams)
most_common = trigram_counts.most_common(1)[0]
```

```python
result = f"{most_common[0][0]} {most_common[0][1]} {most_common[0][2]};{most_common[1]}"
print(result)
```

```python
pred_labels = np.argmax(probas, axis=1)
classes = sorted(set(multiclass_true))
```

```python
results = {}
for cls in classes:
    results[cls] = {"TP": 0, "FP": 0, "TN": 0, "FN": 0}

for i in range(len(multiclass_true)):
    true = multiclass_true[i]
    pred = pred_labels[i]
    for cls in classes:
        if pred == cls and true == cls:
            results[cls]["TP"] += 1
        elif pred == cls and true != cls:
            results[cls]["FP"] += 1
        elif pred != cls and true == cls:
            results[cls]["FN"] += 1
        elif pred != cls and true != cls:
            results[cls]["TN"] += 1
```

```python
precisions = {}
for cls in classes:
    TP = results[cls]["TP"]
    FP = results[cls]["FP"]
    if TP + FP == 0:
        precision = 0.0
    else:
        precision = TP / (TP + FP)
    precisions[cls] = round(precision, 2)
```

```python
best_cls = max(precisions, key=precisions.get)
best_precision = precisions[best_cls]

print(f"{best_cls} → {best_precision}")
```

## Задание 3

Загрузите токенизатор и модель `sentence-transformers/all-MiniLM-L6-v2`. Токенизируйте и рассчитайте векторные представления для двух приведенных ниже текстов. Усредните векторные представления токенов каждого из текстов так, чтобы для каждого из текстов получился вектор размером 384.
Рассчитайте косинусное расстояние между двумя этим векторами, округлите его до 2 знака после запятой и вставьте число в форму.


```python
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.metrics.pairwise import cosine_similarity
```

```python
text1 = "Hounds display remarkable fidelity towards their owners, often brightening days through affectionate gestures."
text2 = "Domestic pups exhibit steadfast allegiance to caregivers, frequently lifting spirits via tender behaviors."
```

```python
model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
inputs1 = tokenizer(text1, return_tensors="pt", truncation=True, padding=True)
inputs2 = tokenizer(text2, return_tensors="pt", truncation=True, padding=True)
```

```python
with torch.no_grad():
    output1 = model(**inputs1).last_hidden_state
    output2 = model(**inputs2).last_hidden_state
```

```python
def mean_pooling(last_hidden, mask):
    mask = mask.unsqueeze(-1).expand(last_hidden.size()).float()
    return (last_hidden * mask).sum(1) / mask.sum(1)

vec1 = mean_pooling(output1, inputs1["attention_mask"])
vec2 = mean_pooling(output2, inputs2["attention_mask"])
```

```python
distance = cosine_similarity(vec1.numpy(), vec2.numpy())[0][0]
print("Близость:", round(distance, 2))
```

```python
from transformers import AutoModelForCausalLM
import torch
```

```python
model_id = "HuggingFaceTB/SmolLM2-135M"

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    load_in_4bit=True,
    device_map="auto"
)

tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
prompt = "I woke up and saw " # hehe пробелы в разных местах промпта дают разные результаты

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

outputs = model.generate(
    **inputs,
    max_new_tokens=40,
    no_repeat_ngram_size=2,
    do_sample=False,
    eos_token_id=tokenizer.eos_token_id
)

generated = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated)
```
